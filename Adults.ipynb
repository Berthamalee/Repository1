```python
import pandas as pd
import numpy as np
```

```python
path = "https://raw.githubusercontent.com/Berthamalee/Repository1/main/datasets/adult.csv"
```

```python
df = pd.read_csv(path)
df
```

```python
pd.pivot_table(df, 
               index= "marital-status",
               )
```

```python
df[["marital-status"]].value_counts()
```

```python
df.columns
```

```python
pd.pivot_table(df,
               index = "marital-status",
               aggfunc = "size", 
               )
```

```python
pd.pivot_table(df, 
               index='marital-status', 
               values=['hours-per-week'], 
               aggfunc = 'mean',
               margins = True,
               columns = 'education'
              )
```

```python
pd.pivot_table(df, 
               index=['education', 'marital-status'],
               values=['hours-per-week', 'age'], 
               aggfunc = 'median',
               margins = True,
               columns = 'sex'
              )
```

```python
pd.pivot_table(df, 
               index='marital-status', 
               values=['hours-per-week'], 
               aggfunc = 'median',
               margins = True,
               columns = 'education'
              )
```
