```python
import pandas as pd
import numpy as np 
import seaborn as sns
import matplotlib as mpl
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
path = "https://raw.githubusercontent.com/Berthamalee/Repository1/main/datasets/antropometria-dataset.csv"
```

```python
df = pd.read_csv(path)
df.tail (2)
```

```python
df.columns
```

```python
df.drop(['intp', 'emb', 'temb', 'tbrazo', 'htension', 'PrimaryLast', 'code_upm',
       'est_dis', 'est_urb', 'est_marg', 'pondef', 'est_var'],
        axis=1, 
        inplace= True
        )
df.tail()
```

```python
df
```

```python
Peso1 = df.groupby(["desc_ent","sexo"])["peso"].median().reset_index()
Peso1
```

```python
Peso2 = Peso1[Peso1["sexo"]==1]
Peso2
```

```python
Peso2["peso"].plot(kind="box")
plt.show;
```

```python
boxplot = df.boxplot(column=["talla"])
plt.title("Box plot Talla")
plt.show ;
```

```python
df["peso"].plot(kind = "line");
```

```python
df["cadera"].plot(kind = "line");
```

```python
sns.lmplot(x= "peso",
           y = 'cadera',
           data = df,
           fit_reg = False, #regression
           hue = 'sexo',
           ); #usando la funcion de lmplot se hace grafica dispersion 
plt.title('Peso vs Cadera', fontsize = 10); #se pueden mezclar las bases de datos
```

```python
boxplot = df.boxplot(column=["cadera"])
plt.title("Box plot cadera")
plt.show ;
```

```python
boxplot = df.boxplot(column=["cintura"])
plt.title("Box plot cintura")
plt.show ;
```

```python
Cadera1 = df.groupby(["desc_ent","sexo"])["cadera"].median().reset_index()
Cadera1
```

```python
Cadera1a = Cadera1[Cadera1["sexo"]==2]
Cadera1a
```

```python
plt.plot(Cadera1a["cadera"])
plt.title("Cadera Line Plot")
plt.xlabel("Inidencias")
plt.ylabel("Cadera")
plt.show
```

```python
Ropa1 = df.groupby(["desc_ent","sexo"])["ropa"].sum().reset_index()
Ropa1
```

```python
Ropa1a = Ropa1[Ropa1["sexo"]==1]
Ropa1a
```

```python
Ropa1b = Ropa1a.sort_values("ropa", ascending = False)
Ropa1b
```

```python
x= Ropa1b.head(10)
x
```

```python
x["ropa"].plot(kind="bar")
plt.show;
```
