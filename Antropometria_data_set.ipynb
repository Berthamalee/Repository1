```python
import pandas as pd
import numpy as np
```

```python
path = "https://raw.githubusercontent.com/Berthamalee/Repository1/main/datasets/antropometria-dataset-2.csv"
```

```python
df = pd.read_csv(path)
df
```

```python
df.columns 
```

```python
pd.pivot_table(df, 
               index = "entidad",
               )
```

```python
pd.pivot_table(df,
               index= "desc_ent",
               values ="peso",
               )
```

```python
pd.pivot_table(df, 
               index = "sexo",
               values = "peso",
               aggfunc = "mean",
               columns = "desc_ent",
               )
```

```python
pd.pivot_table(df, 
               index = "desc_ent",
               values = "peso",
               aggfunc = "median",
               columns = "sexo",
               )
```
