```python
import pandas as pd 
import numpy as np
import seaborn as sns 
import matplotlib as mpl 
import matplotlib.pyplot as plt
import statsmodels.api as sm
from sklearn import linear_model
```

```python
path = '/content/drive/MyDrive/Challenge Agua/challenge 2.xlsx'
```

```python
df = pd.read_excel(path)
df.head(2)
```

```python
#df.isnull().any()
```

```python
columns = df.loc[:,('Movilidad','VinosYLicores')].fillna(0)
```

```python
y = df.loc[:,('Movilidad','VinosYLicores')].fillna(0)
XX = df[["Año 2020"]]
from sklearn.model_selection import train_test_split as t_split
from sklearn.metrics import r2_score as r2
[x_train,x_test,y_train,y_test]=t_split(XX,y,test_size=0.2,random_state=0)
print(np.size(XX))
print(np.size(y))
```

```python
lm = linear_model.LinearRegression()
lm.fit(XX, y)
y_prediction_train=lm.predict(x_train)
score=r2(y_train,y_prediction_train)
print(score)
```

```python
lm = linear_model.LinearRegression()
model_lm = lm.fit(XX, y)
```

```python
print("The intercept is ", model_lm.intercept_)
print("The coefficient are ", model_lm.coef_)
print("The r^2 value is ", model_lm.score(XX, y))
```
